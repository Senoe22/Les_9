```python
# Задание 1
log = pd.read_csv('visit_log.csv', sep = ';')

# Для проверки - распределение данных в стодбце "traffic_source"
#print(log.traffic_source.value_counts())

# Yandex и google = organic в столбце "source_type"
log.loc[(log.traffic_source == 'yandex') | (log.traffic_source =='google'), 'source_type'] = "organic"

# paid и email = ad для Russia в столбце "source_type"
log.loc[(log.region.str.contains('Russia')) & ((log.traffic_source == 'paid') | (log.traffic_source =='email')), 'source_type'] = "ad"

# paid и email = other для не Russia в столбце "source_type"
log.loc[(log.region != 'Russia') & ((log.traffic_source == 'paid') | (log.traffic_source =='email')), 'source_type'] = "other"

# Для остальных (пустых) значений в столбце "source_type" присваиваем значнения из столбца "traffic_source"
log.source_type.fillna(log.traffic_source, inplace = True)
log.head()
```

```python
# Задача 3
stats = pd.read_csv('ml-latest-small/ratings.csv')
#stats['date'] = pd.to_datetime(stats['timestamp'], unit = 's')

# Преобразуем таблицу в сводную с итогом по сумме выставленных оценок
pivot_stats = stats.pivot_table(index='userId', columns = 'rating', values = 'timestamp', aggfunc = 'count', margins = True)
#print(pivot_stats.head())

# Добавляем столбец "Clients_typy" с указанием "Active" для userId с общим количеством выставленных 
#оценок 100 и более
pivot_stats.loc[pivot_stats.All >= 100, 'Clients_type'] = "Active"
#pivot_stats.head()

# Переводим userId в индекс для stats(чтобы можно было воспользоваться join)
new_stats = stats.set_index('userId')

# К new_stats присоединяем столбец "Clients_type"
joined = new_stats.join(pivot_stats.Clients_type)
joined.head()

# Переводим userId из индекса (чтобы можно было воспользоваться pivot)
index_joined = joined.reset_index()

# Сводная таблица с указанием мин.времени
joined_pivot_min = index_joined.pivot_table(index='userId', columns = 'Clients_type', values = 'timestamp', aggfunc = 'min', margins = True )
joined_pivot_min.head()

# Сводная таблица с указанием макс.времени
joined_pivot_max = index_joined.pivot_table(index='userId', columns = 'Clients_type', values = 'timestamp', aggfunc = 'max', margins = True )
# Переименованние названий столбцов, чтобы они не совпадали с проедыдущей таблицей (мин)
joined_pivot_max_rename = joined_pivot_max.rename(columns = {'Active': 'Max_time', 'All' : '222'})

# Объединение таблиц с мин и макс временем клиентов с количеством оценок 100 и более
joined_time = joined_pivot_min.join(joined_pivot_max_rename)

# Удаление ненужных столбцов
joined_time.drop(['All', '222'], axis = 'columns', inplace = True)

# Переименование названий столбцов
final = joined_time.rename(columns = {'Active': 'Min_time', 'Max_time' : 'Max_time'})

# Перевод времени в привычный формат дат
final['Min_time'] = pd.to_datetime(final['Min_time'], unit = 's')
final['Max_time'] = pd.to_datetime(final['Max_time'], unit = 's')
final['diff'] = final['Max_time'] - final['Min_time']
print('Среднее время жизни клиентов, у которых 100 и более оценок - ', final['diff'].mean())
```

```python
# Задача 4
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)

#Переводим client_Id в индекс (чтобы можно было воспользоваться join)
new_rzd = rzd.set_index('client_id')
new_auto = auto.set_index('client_id')
new_air = air.set_index('client_id')
new_client_base = client_base.set_index('client_id')

# К rzd добавляем столбец с выручкой auto 
rzd_auto = new_rzd.join(new_auto.auto_revenue, how = 'outer')

# К rzd и auto добавляем столбец с выручкой air
rzd_auto_air = rzd_auto.join(new_air.air_revenue, how = 'outer')

# К rzd, auto, air добавляем столбец с адресами клиентов
rzd_auto_air_client_base = rzd_auto_air.join(new_client_base.address, how = 'outer')

print('Таблица 1')
print(rzd_auto_air.head(15))

print('Таблица 2')
print(rzd_auto_air_client_base.head(15))
```
